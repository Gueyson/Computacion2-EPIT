```python
import numpy as np
import matplotlib.pyplot as plt

```

```python
# Definir la función
def funcion_exponencial(x):
    return np.exp(-x**2)

x = np.linspace(-1, 3, 100)
y = funcion_exponencial(x)

fig, ax = plt.subplots()

ax.plot(x, y)

plt.gca().invert_yaxis()

ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_title('FUNCIÓN EXPONENCIAL DE LA FORMA invertida : f(x) = e^(-x^2)')

plt.show()
```

### Hemos notado que en el ejercicio de arriba, no es lo que nos pide. La gráfica es identíca porque el eje y está invertido.

```python
def funcion_exponencial(x):
    return np.exp(-x**2)

def funcion_g(x):
    return 1 - funcion_exponencial(x)

x = np.linspace(-1, 3, 100)

y = funcion_g(x)

fig, ax = plt.subplots()

ax.plot(x, y)

ax.set_xlabel('x')
ax.set_ylabel('g(x)')
ax.set_title('Gráfica de la función g(x) = 1 - f(x)')

plt.show()
```

### Sin embargo, ahora usamos conocimiento matemético para analizar el comportamiento de la función g(x)=1-f(x)
