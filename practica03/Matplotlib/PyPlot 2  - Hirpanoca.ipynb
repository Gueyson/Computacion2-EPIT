## GRÁFICAS UTILIZANDO NUMPY:

###  Importamos los paquetes necesarios:

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

%matplotlib inline
```

## Configuración del estilo del gráfico y del tamaño de la fuente

###  TIPO 1

```python
sns.set_style('whitegrid') # darkgrid, whitegrid, dark, white and ticks
plt.rc('axes', titlesize=18) #Tamaño del titulo de los ejes
plt.rc('axes', labelsize=14) #tamaño de las etiquetas x , y 
plt.rc('xtick', labelsize=13)# tamaño de las etiquetas de los ticks
plt.rc('ytick', labelsize=13)# tamaño de las etiquetas de los ticks
plt.rc('legend', fontsize=13)# tamaño de las fuentes de la leyenda
plt.rc('font', size=13)      # controla el tamaño del texto por defecto
```

###  TIPO 2

```python
#plt.style.use('seaborn-whitegrid')
plt.rc('axes', titlesize=18)  # tamaño del título de los ejes
plt.rc('axes', labelsize=14)  # tamaño de las etiquetas x , y
plt.rc('xtick', labelsize=13) # tamaño de las etiquetas de los ticks
plt.rc('ytick', labelsize=13) # tamaño de las etiquetas de los ticks
plt.rc('legend', fontsize=13) # tamaño de las fuentes de la leyenda
plt.rc('font', size=13)       # controla el tamaño del texto por defecto
```

```python
x = np.linspace(0, 15, 100)
y = np.sin(x)
plt.plot(x, y)
plt.show()
```

```python
plt.figure(figsize=(15, 7), dpi =80)
plt.plot(x,y)
plt.title('f(x) -> seno(x)', fontsize=20)
plt.xlabel('x', fontsize=20)
plt.ylabel('seno(x)', fontsize=20)
plt.xlim(0,10)
plt.ylim(-1.1,1.1)
plt.savefig('funcion_seno1.png')
plt.show()
```

```python
y2 = np.cos(x)
fig, (ax1, ax2) = plt.subplots(nrows=1, ncols=2,
figsize=(14,5))
ax1.scatter(x=x, y=y, marker='.', c='r', edgecolor='b')
ax1.set_title('Scatter: x versus sen(x)')
ax1.set_xlabel('$x$')
ax1.set_ylabel('$sen(x)$')
ax2.plot(x, y2, color='r',
         label=('cos(x)'))
ax2.legend(loc='best')
ax2.set_title('Grafico lineal: x versus cos(x)')
ax2.yaxis.tick_right()
ax2.set_xlabel('$x$')
ax2.set_ylabel('$cos(x)$')

plt.show()
```

## <font color=#33A5FF>****3. Gráfico de Barras****<tfont>

**PREPARACIÓN DE DATOS**

```python
df[['year','age']]
```

```python
df_group = df.groupby(['year'])['age'].count().reset_index()
df_group
```

```python
df_group.columns = ['year','cantidad']
df_group
```

```python
df_group = df_group.sort_values(by=['cantidad'], ascending=False)
df_group
```

```python
plt.figure(figsize=(15, 7), dpi =80)
plt.bar(df_group['year'], df_group['cantidad'], color='orange')
plt.xticks(rotation=70)
plt.xlabel('AÑO')
plt.ylabel('N°DE PERSONAS')
plt.title('PERSONAS INTERVENIDAS POR AÑO')
plt.show()
```
