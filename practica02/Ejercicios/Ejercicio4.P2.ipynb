```python
#PRIMERO IMPORTAMOS LIBRERÍAS
import numpy as np
import scipy
#MINILIBRERÍAS MÁS ESPECÍFICAS
from numpy import matrix
from scipy import linalg
def MENU_OPERACIONES_CASIO():
    print ('\n')
    titulo = ("BUEN DÍA QUÉ DESEA HACER HOY")
    print (titulo.center(50,'+'))
    print ("Selecciona el NÚMERO de la operación")
    print ("\t1 Multiplicacion")
    print ("\t2 Inversa de la matriz")
    print ("\t3 Sistema de ecuaciones lineales")
    print ("\t9 Ninguna operación por ahora")
    
def mat():
    filas = int(input("Número de filas matriz 1: "))
    columnas = int(input("Número de columnas matriz 1: "))
    matriz = []
    for i in range(filas):
        matriz.append([])
        for j in range(columnas):
            valor = float(input("Fila {}, Columna {} : ".format(i+1, j+1)))
            matriz[i].append(valor)
    for fila in matriz:
        print("[", end=" ")
        for elemento in fila:
            print("{:8.2f}".format(elemento), end=" ")
        print("]")
    return matriz

def mat2():
    filas = int(input("Número de filas matriz 2: "))
    columnas = int(input("Número de columnas matriz 2: "))
    matriz1 = []
    for i in range(filas):
        matriz1.append([])
        for j in range(columnas):
            valor = float(input("Fila {}, Columna {} : ".format(i+1, j+1)))
            matriz1[i].append(valor)
    for fila in matriz1:
        print("[", end=" ")
        for elemento in fila:
            print("{:8.2f}".format(elemento), end=" ")
        print("]")
    return matriz1
        

        
def validar():
    while True: 
        try:
            # Mostramos el menu
            MENU_OPERACIONES_CASIO()
            # solicituamos una opción al usuario
            print ('\n')
            opcionMenu = input("Selecciona el número de la operación: ")
            print ('\n')
            if opcionMenu=="1":
                print ('\n')
                A = matrix(mat()) 
                B = matrix(mat2()) 
                print(A*B)
            elif opcionMenu=="2":
                print ('\n')
                print("Inversa de la Matriz1")
                A = matrix(mat()) 
                print(linalg.inv(A))
                input("Has pulsado la opción 2...\npulsa una tecla para continuar")
            elif opcionMenu=="3":
                print ('\n')
                A = matrix(mat())
                B = matrix(mat2()) 
                print(linalg.solve(A,B))
                input("Has pulsado la opción 3...\npulsa una tecla para continuar") 
            elif opcionMenu=="9":
                break
            else:
                print ("")
                input("No has pulsado ninguna opción correcta...\npulsa un número 1 2 3 o 9 para continuar")
        except ZeroDivisionError as e:
            print('DIVISIÓN ENTRE CERO NOOO0, NUNCA')
        except ValueError:
            print('La entrada a debe ser una matriz cuadrada...')
    return opcionMenu


                  
                  
if __name__ == '__main__':
    validar()
    print('\tFINALIZO EL PROGRAMA!!!!!')
```
